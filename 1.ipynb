## Identify Dissatisfied Employees

TAFE DATASET

```python
 data_tafe_updated.head()
```

```python
 data_tafe_updated['Contributing Factors. Dissatisfaction'].value_counts()
```

```python
 data_tafe_updated['Contributing Factors. Job Dissatisfaction'].value_counts()
```

```python
def update_vals(element):
    if element == '-':
        return False
    elif pd.isnull(element):
        return np.nan
    else:
        return True
  
data_tafe_updated[['Contributing Factors. Job Dissatisfaction', 'Contributing Factors. Dissatisfaction']] = (
data_tafe_updated[['Contributing Factors. Job Dissatisfaction', 'Contributing Factors. Dissatisfaction']]
    .applymap(update_vals))

#True: Employee was dissatisfied over one of more things, leading to their resignation
#False: Employee resigned for a reason other than dissatisfaction   
data_tafe_updated['dissatisfied'] = data_tafe_updated[['Contributing Factors. Job Dissatisfaction', 'Contributing Factors. Dissatisfaction']].any(1)
```

## Combine the Data

```python
dete_resignations_up['institute'] = 'DETE'
tafe_resignations_up['institute'] = 'TAFE'
```

```python
combined = pd.concat([dete_resignations_up, tafe_resignations_up], ignore_index=True)

combined.head()
```

```python
combined.shape
```

```python
combined_updated = combined.dropna(thresh=500, axis = 1)

combined_updated.notnull().sum().sort_values()
```

## Clean the Service Column

```python
combined_updated["institute_service"].value_counts()
```

```python
extracted_years = (combined_updated['institute_service']
                  .astype(str).str.extract(r'(\d+)')
                  .astype('float'))

combined_updated['institute_service_up'] = extracted_years

combined_updated['institute_service_up'].value_counts()
```

```python
def career_stage(element):
    if element < 3:
        return 'New'
    elif 3 <= element < 7:
        return 'Experienced'
    elif 7 <= element < 11:
        return 'Established'
    elif pd.isnull(element):
        return np.nan
    else:
        return 'Veteran'
    
# Apply this function to `institute_service_up` and store the results in `service_cat`.

combined_updated['service_cat'] = combined_updated['institute_service_up'].apply(career_stage)

combined_updated['service_cat'].value_counts()
```
