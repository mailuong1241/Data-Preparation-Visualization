```python
#ex1
import pandas as pd
import seaborn as sns
#a
appl = pd.read_csv( "aapl.csv")
amzn = pd.read_csv("amzn.csv")
fb = pd.read_csv("fb.csv")
goog = pd.read_csv("goog.csv")
nflx = pd.read_csv("nflx.csv")
```

```python
#b
appl['ticker'] = 'appl'
amzn['ticker'] = 'amzn'
fb['ticker'] = 'fb'
goog['ticker'] = 'goog'
nflx['ticker'] = 'nflx'
faang = pd.concat([appl,amzn,fb,goog,nflx])
faang
```

```python
#c
faang.to_csv('faang.csv')
```

```python
#d
df = pd.read_csv('faang.csv', index_col=[0])
df
```

```python
#e
df = df.assign(date = lambda x: pd.to_datetime(x.date),volume = lambda x: x.volume.astype(int))
df= df.sort_values(by=['date','ticker'],ascending=[True,True])
df
```

```python
#f
lowest = df.sort_values(by='volume')
lowest.head(7)
```

```python
#ex2
#a
covid19 = pd.read_csv('covid19_cases.csv')
covid19.head()
```

```python
#b
covid19['Date'] = pd.to_datetime(covid19.dateRep)
covid19.head()
```

```python
#c
covid19.set_index('Date', inplace= True)
covid19.sort_values(by='Date')
covid19.head()
```

```python
#d
covid19[covid19.countriesAndTerritories == "United_States_of_America"]
covid19.replace({'United_States_of_America':"US","United_ Kingdom":"UK"},inplace=True)
covid19[covid19.countriesAndTerritories=='US']
```
